# Wykres pudełkowy

Wykres pudełkowy (box plot) to graficzna prezentacja rozkładu danych jednowymiarowych, która opiera się na pięciu kluczowych wartościach:

Minimum (Min) – najmniejsza wartość danych,

Pierwszy kwartyl (Q1) – wartość, poniżej której znajduje się 25% danych,

Mediana (Me) – środkowa wartość danych dzieląca je na połowy,

Trzeci kwartyl (Q3) – wartość, poniżej której znajduje się 75% danych,

Maksimum (Max) – największa wartość danych.

Interpretacja poszczególnych elementów wykresu:
Pudełko: rozciąga się od Q1 do Q3 i obejmuje 50% środkowych danych. Długość pudełka to rozstęp międzykwartylowy (IQR), czyli Q3 – Q1.

Linia w środku pudełka: oznacza medianę – wartość środkową rozkładu.

Wąsy: linie rozciągające się od pudełka do wartości minimalnej i maksymalnej (lub do wartości granicznych definiowanych przez 1,5 × IQR). Pokazują rozpiętość typowych danych.

Punkty poza wąsami (outliers): wartości ekstremalne, które znacznie odbiegają od pozostałych obserwacji.

Co można odczytać z wykresu pudełkowego?
Rozkład danych i ich rozproszenie (im dłuższy wykres, tym większa zmienność).

Położenie mediany i symetrię rozkładu (jeśli mediana jest blisko środka pudełka, rozkład jest symetryczny).

Obecność wartości odstających.

Ewentualną asymetrię danych (np. wydłużone wąsy z jednej strony wskazują na prawostronną lub lewostronną asymetrię).

Wykres pudełkowy jest powszechnie używany do porównywania rozkładów w różnych grupach, zwłaszcza przy analizie statystycznej i eksploracji danyc

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bins = range(0, int(df['Calories'].max()) + 100, 100)
df['Calories_bin'] = pd.cut(df['Calories'], bins=bins, right=False)

g = sns.catplot(
    data=df,
    x="Calories_bin", y="Pulse",
    kind="box",
    height=5,
    aspect=2
)

print(df.describe())
plt.xticks(rotation=60)  # obrót etykiet dla lepszej czytelności
plt.show()
```

# wykres wiolinowy

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bins = range(0, int(df['Calories'].max()) + 100, 100)
df['Calories_bin'] = pd.cut(df['Calories'], bins=bins, right=False)

g = sns.catplot(
    data=df,
    x="Calories_bin", y="Pulse",
    kind="violin",
    height=5,
    aspect=2
)

plt.xticks(rotation=45)  # obrót etykiet dla lepszej czytelności
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bins = range(0, int(df['Calories'].max()) + 100, 100)
df['Calories_bin'] = pd.cut(df['Calories'], bins=bins, right=False)

g = sns.catplot(
    data=df,
    x="Calories_bin", y="Pulse",
    kind="violin",
    hue="Calories_bin",
    height=5,
    aspect=2
)

plt.xticks(rotation=45)  # obrót etykiet dla lepszej czytelności
plt.show()
```
