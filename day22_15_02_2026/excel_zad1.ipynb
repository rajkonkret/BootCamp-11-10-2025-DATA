#### Dodanie kolumny - suma sprzedaży

```python
row_position = 2
col_position = 7

total_sales = (
        (ws.cell(row=row_position, column=col_position).value) +
        (ws.cell(row=row_position, column=col_position + 1).value) +
        (ws.cell(row=row_position, column=col_position + 2).value) +
        (ws.cell(row=row_position, column=col_position + 3).value)

)

ws.cell(row=2, column=11).value = total_sales
wb.save("video2.xlsx")
```

```python
row_position = 1

for i in range(1, ws.max_row):
    row_position += 1
    NA_Sales = ws.cell(row=row_position, column=7).value
    EU_Sales = ws.cell(row=row_position, column=8).value
    JP_Sales = ws.cell(row=row_position, column=9).value
    Other_Sales = ws.cell(row=row_position, column=10).value

    total_sales = NA_Sales + EU_Sales + JP_Sales + Other_Sales
    ws.cell(row=row_position, column=11).value = total_sales

wb.save('video2.xlsx')
```

#### Wykres słupkowy

```python
from openpyxl.chart import Reference
```

```python
values = Reference(
    ws,
    min_col=2,
    max_col=2,
    min_row=1,
    max_row=13
)
```

```python
cats = Reference(
    ws,
    min_col=1,
    max_col=1,
    min_row=2,
    max_row=13
)
```

```python
from openpyxl.chart import BarChart
```

```python
chart = BarChart()
chart.add_data(values, titles_from_data=True)
chart.set_categories(cats)

chart.title = 'Total Sales'
chart.x_axis.title = "Genre"
chart.y_axis.title = "Total Sales by Genre"

ws.add_chart(chart, 'D2')
wb.save('video2.xlsx')
```

```python
wb.sheetnames
```

```python
# 'Breakdown of Sales by Genre'
```

```python
ws = wb['Breakdown of Sales by Genre']
values = Reference(
    ws,
    min_col=2,
    max_col=5,
    min_row=1,
    max_row=13
)

cats = Reference(
    ws,
    min_col=1,
    max_col=1,
    min_row=2,
    max_row=13
)

chart = BarChart()
chart.add_data(values, titles_from_data=True)
chart.set_categories(cats)

chart.title = 'Sales Breakdown'
chart.x_axis.title = "Genre"
chart.y_axis.title = "Breakdown of Sales by Genre"

ws.add_chart(chart, 'H2')
wb.save('video2.xlsx')
```

Wykres liniowy

```python
values = Reference(
    ws,
    min_col=2,
    max_col=6,
    min_row=1,
    max_row=40
)

cats = Reference(
    ws,
    min_col=1,
    max_col=1,
    min_row=2,
    max_row=40
)
```

Formatowanie warunkowe

```python
from openpyxl.formatting.rule import CellIsRule
```

```python
fill = PatternFill(
    start_color='90EE90',
    end_color="90EE90",
    fill_type="solid"
)
```

expand = {">": "greaterThan", ">=": "greaterThanOrEqual", "<": "lessThan", "<=": "lessThanOrEqual",
              "=": "equal", "==": "equal", "!=": "notEqual"}

```python
ws.conditional_formatting.add(
    'G2:K16600',
    CellIsRule(
        operator='greaterThan',
        formula=[8],
        fill=fill
    )
)
```

```python
wb.save('video2.xlsx')
```

```python
fill = PatternFill(
    start_color="770E77",
    end_color="770EFF",
    fill_type="darkUp"
)
```

```python
ws.conditional_formatting.add(
    "G2:K16600",
    CellIsRule(
        operator="lessThan",
        formula=[4],
        fill=fill,
        font=Font(color='FF0000')
    )
)
```
