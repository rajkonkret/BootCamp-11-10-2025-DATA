```python
import openpyxl

wb = openpyxl.load_workbook("videogamesales.xlsx")
wb
```

```python
ws = wb.active
ws
```

```python
ws = wb["vgsales"]  # jawne wybranie arkusza
ws
```

```python
print("total number of rows:", ws.max_row)
```

```python
print("Total number of columns:", ws.max_column)
```

```python
print("The value in cell A1 is:", ws['A1'])  # <Cell 'vgsales'.A1>
print("The value in cell A1 is:", ws['A1'].value)  # Rank
```

```python
values = [ws.cell(row=1, column=i).value for i in range(1, ws.max_column + 1)]
# +1 bo zbiór otwart a excel liczy od 1
print(values)
```
